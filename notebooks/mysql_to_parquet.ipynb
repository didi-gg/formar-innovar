```python
# Carpeta base de salida
output_base_dir = "./parquet_exports/"

# Crear la carpeta base si no existe
os.makedirs(output_base_dir, exist_ok=True)

try:
    conn = pymysql.connect(**db_config)
    print("Conexión exitosa a la base de datos.")

    with conn.cursor() as cursor:
        for group, tables in tables_by_group.items():
            # Crear directorio para el grupo
            group_dir = os.path.join(output_base_dir, group)
            os.makedirs(group_dir, exist_ok=True)
            
            for table in tables:
                print(f"Cargando tabla: {table}")
                
                # Leer la tabla
                cursor.execute(f"SELECT * FROM {table}")
                rows = cursor.fetchall()
                columns = [col[0] for col in cursor.description]
                
                # Convertir a PyArrow Table
                arrow_table = pa.Table.from_pydict(
                    {col: [row[i] for row in rows] for i, col in enumerate(columns)}
                )
                
                # Ruta del archivo Parquet
                output_file = os.path.join(group_dir, f"{table}.parquet")
                
                # Exportar la tabla a Parquet
                print(f"Exportando tabla {table} a {output_file}...")
                pq.write_table(arrow_table, output_file, compression="snappy")
                print(f"Tabla {table} exportada exitosamente.")

except pymysql.MySQLError as e:
    print(f"Error conectando a la base de datos: {e}")

finally:
    if conn:
        conn.close()
        print("Conexión cerrada.")
```

```python
import duckdb

# Rutas de los archivos Parquet
user_parquet = "./parquet_exports/Users/mdlvf_user.parquet"
user_info_data_parquet = "./parquet_exports/Users/mdlvf_user_info_data.parquet"

# Crear conexión DuckDB
con = duckdb.connect()

# Registrar los archivos Parquet como tablas temporales en DuckDB
con.execute(f"CREATE TABLE mdlvf_user AS SELECT * FROM parquet_scan('{user_parquet}')")
con.execute(f"CREATE TABLE mdlvf_user_info_data AS SELECT * FROM parquet_scan('{user_info_data_parquet}')")

# Ejecutar la consulta
query = """
SELECT 
    u.id AS userid,
    u.idnumber,
    u.firstname,
    u.lastname,
    u.city,
    u.firstaccess,
    u.timecreated,
    u.lastaccess,
    u.lastlogin,
    u.currentlogin,
    uid.data AS role_data
FROM 
    mdlvf_user u
JOIN 
    mdlvf_user_info_data uid ON u.id = uid.userid
WHERE 
    uid.data = 'Estudiante' AND u.idnumber <> '';
"""

result = con.execute(query).fetchdf()  # Devuelve un DataFrame de pandas

# Mostrar los resultados
print(result)

# Opcional: guardar los resultados en un nuevo archivo Parquet
result.to_parquet("filtered_users.parquet", index=False)
```
